### Препроцессинг датасета с шутками

```python
from datasets import load_dataset
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
import json
```

```python
with open(r'D:\Chat-bot\jokes\dataset.json', 'r', encoding='utf-8') as f:
    text = f.read().split('\n')
```

```python
text
```

```python
def preprocess_string(x:text):
    x = x.split(' [')[1].replace(']}', '').split('",')
    result = []
    for el in x:
        result.append(el.replace('"', ''))
    return ';;;'.join(result)

df = pd.Series(text)
df = df.apply(lambda x: preprocess_string(x))
df[0]
```

```python
df.to_csv(r'D:\Chat-bot\jokes\df.csv', sep='\t', encoding='utf-8')
```
